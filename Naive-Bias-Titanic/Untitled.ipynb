# Titanic_Naive_Bayes

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
df=pd.read_csv(r"G:\DATA-SCIENCE\titanicNB.csv")
df
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
df
```

```python
inputs=df.drop('Survived',axis='columns')
inputs
```

```python
target=df.Survived
target
```

```python
dummies=pd.get_dummies(inputs.Sex)
dummies
```

```python
inputs2=pd.concat([inputs,dummies],axis='columns')
inputs2
```

```python
inputs2.drop(['Sex','male'],axis='columns',inplace=True)
inputs2
```

```python
inputs2.columns[inputs2.isna().any()]
```

```python
mean=inputs2.Age.mean()
mean
```

```python
inputs2.Age=inputs2.Age.fillna(mean)
inputs2
```

```python
X_train,X_test,y_train,y_test=train_test_split(inputs2,target,test_size=0.2)
```

```python
model=GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.predict(X_test)
```

```python
model.predict_proba(X_test)
```

cross-validation

```python
from sklearn.model_selection import cross_val_score
cross_val_score(model,X_train,y_train,cv=5)
```
